```python
#gerando qrcodes
import qrcode


for i in range(len(base_html)):
    link = f'https://guii-barbosa.github.io/candidatos/{base_html["Inscrição"][i]}/{base_html["Inscrição"][i]}.html'
    qrimg = qrcode.make(link)
    qrimg.save(f'./qrCodes/{base_html["Inscrição"][i]}.png')

```
